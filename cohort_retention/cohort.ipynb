# Cohort Retention — когортный анализ

```python
import pandas as pd
import numpy as np

# Пример синтетических логов: user_id, first_month, month_active
users = 500
rng = np.random.default_rng(1)
first_month = rng.integers(1, 4, size=users)  # когорты 1..3
rows = []
for u in range(users):
    fm = first_month[u]
    # активность в течение 6 месяцев с вероятностью затухания
    for m in range(fm, 7):
        if rng.random() < max(0.8 - 0.12*(m-fm), 0.05):
            rows.append((u, fm, m))
            
df = pd.DataFrame(rows, columns=["user_id","cohort","month"])

# Пивот удержания
cohort_size = df.groupby("cohort")["user_id"].nunique()
ret = df.groupby(["cohort","month"])["user_id"].nunique().unstack(fill_value=0)
ret = ret.div(cohort_size, axis=0).round(3)
ret
```

Матрица показывает долю активных пользователей по месяцам с момента привлечения (по строкам — когорты).
